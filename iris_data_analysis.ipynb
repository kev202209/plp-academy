```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import datasets

```

# Task 1: Load and Explore the Dataset

```python


iris = datasets.load_iris()
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
iris_df['species'] = pd.Categorical.from_codes(iris.target, iris.target_names)

print("First few rows of the dataset:")
print(iris_df.head())

    
```

```python
print("\nDataset information:")
print(iris_df.info())
```

```python
# Check for missing values
print("\nMissing values:")
print(iris_df.isnull().sum())

```

### No missing values in the Iris dataset, so no cleaning needed.

# Task 2: Basic Data Analysis

```python
print("\nBasic statistics of numerical columns:")
print(iris_df.describe())
```

```python
print("\nAverage petal length per species:")
print(iris_df.groupby('species')['petal length (cm)'].mean())
```

### Observations:
 - Iris-setosa has the smallest petal length and width.
 - Iris-virginica has the largest petal length and width.
 - Sepal length and width show some variation across species.

# Task 3: Data Visualization

```python
# Bar chart: Average petal length per species
plt.figure(figsize=(8, 6))
sns.barplot(x='species', y='petal length (cm)', data=iris_df)
plt.title('Average Petal Length per Species')
plt.xlabel('Species')
plt.ylabel('Petal Length (cm)')
plt.show()
```

```python
# Histogram: Petal width distribution
plt.figure(figsize=(8, 6))
sns.histplot(iris_df['petal width (cm)'], kde=True)
plt.title('Distribution of Petal Width')
plt.xlabel('Petal Width (cm)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Scatter plot: Sepal length vs. petal length
plt.figure(figsize=(8, 6))
sns.scatterplot(x='sepal length (cm)', y='petal length (cm)', hue='species', data=iris_df)
plt.title('Sepal Length vs. Petal Length')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Petal Length (cm)')
plt.legend(title='Species')
plt.show()
```

```python
# Boxplot of sepal width per species
plt.figure(figsize=(10, 6))
sns.boxplot(x='species', y='sepal width (cm)', data=iris_df)
plt.title('Sepal Width Distribution by Species')
plt.xlabel('Species')
plt.ylabel('Sepal Width (cm)')
plt.show()
```
